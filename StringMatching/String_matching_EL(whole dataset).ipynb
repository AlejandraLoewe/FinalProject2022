```python
#compare every term in the dataset gain
words = df['Term'].unique()
text = df['General SNOMED Label'].unique()
rows = []
final = []
count = 1
for word in words:
    for tex in text:
        #print(word, tex)
        count += 1
        print("count is " , count)
        diff_jr = lev.jaro_winkler(word, tex)
        diff_lev = lev.distance(word, tex)
        #print(diff)
        #array += [diff_jr]
        #array2 += [diff_lev]
        #print()
        #print(word, tex, diff)
        rows.append([word, tex, diff_jr, diff_lev])
        data = pd.DataFrame(rows, columns = ["Entity", "SNOMED ID", "J-W", "Lev" ])
        #data['JW'] = diff
```

```python
data
```

```python
#Create a new colums with the maximum value obtained when comparing to all strings
data['Max'] = data.groupby('Entity')['J-W'].transform('max')

#Create a colum to check if the jaro measurement is the same in the row
data['True'] = (data['Max'] == data['J-W'])

#keep only one value for each, keeping only the one with the highest value
data_filtered = data[(data['True']==True)]

#drop the entity duplicates
data = data_filtered.drop_duplicates(subset=['Entity'])
data
```

```python
#add the original dataset to get the accuracy
data_to_add = data_original[:30]
data_to_add = data_to_add[[
    "Term",
    "General SNOMED Label"
]
]
data_to_add = data_to_add.drop_duplicates()
data_to_add
```

```python
accuracy = pd.merge(data, data_to_add, left_on ="Entity", right_on='Term' )
#create another column for the match of accuracy
accuracy['Result'] = (accuracy['SNOMED ID'] == accuracy['General SNOMED Label'])
accuracy
```
